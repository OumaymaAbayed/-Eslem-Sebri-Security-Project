```python
import pandas as pd
df = pd.read_csv('tayara_vehicles_raw_data.csv', encoding='cp1252')
```

```python
# Drop unnecessary columns
df = df.drop(columns=['Location','Puissance fiscale', 'Cylindrée moto', 'Livraison'])

# Rename columns
df = df.rename(columns={
    'Kilométrage': 'Mileage',
    'Couleur du véhicule': 'Color',
    'Etat du véhicule': 'State',
    'Boite': 'Gearbox',
    'Année': 'Year',
    'Cylindrée ': 'Engine_cap',
    'Marque': 'Brand',
    'Modèle': 'Model',
    'Type de carrosserie': 'Body_type',
    'Carburant': 'Fuel'
})
```

```python
import numpy as np

# Modify 'Price': remove 'DT', convert to numeric, filter range
df['Price'] = df['Price'].str.replace('DT', '', regex=False).str.replace(' ', '').str.strip()
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Replace prices above 500,000 with NaN
df.loc[df['Price'] > 500000, 'Price'] = np.nan

# Replace all NaN prices with 'upon request'
df['Price'] = df['Price'].fillna('upon request')
```

```python
# Fill missing values in 'Color' with 'Autre'
df['Color'] = df['Color'].fillna('Autre')
```
